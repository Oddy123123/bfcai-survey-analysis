```python
from analyze import *

import tkinter as tk
from tkinter import *
from tkinter import filedialog
from tkinter import messagebox
import tkinter.font as font
import traceback
import os

last_dir = os.getcwd()

def selectFile(course_type):
    global last_dir
    file_name = filedialog.askopenfilename(initialdir=last_dir, title="Select file")
    if not file_name:
        return
    try:
        last_dir = os.path.dirname(file_name)
        analyze(file_name, course_type)
        messagebox.showinfo("Information", "Done")
    except:
        messagebox.showerror("Error", "Failed")
        print(traceback.format_exc())
        
def set_window_size(window, width, height):
    screenwidth = window.winfo_screenwidth()
    screenheight = window.winfo_screenheight()
    x = int((screenwidth - width)/2)
    y = int((screenheight - height)/2)
    window.geometry("{}x{}+{}+{}".format(width, height, x, y))
    
window = tk.Tk()
set_window_size(window, 400, 300)

container = tk.Frame(window)
container.place(relx=0.5, rely=0.5, anchor=CENTER)

lbl_course_type = tk.Label(container, text="اختر نوع المقرر")
lbl_course_type.pack()
lbl_course_type['font'] = font.Font(size=15)
lbl_course_type['fg'] = "Blue"

course_type_var = IntVar()
course_type_var.set(1)

radio1 = Radiobutton(container, text="محاضرة فقط", variable=course_type_var, value=1)
radio1['font'] = font.Font(size=15)
radio1.pack(anchor = W)

radio2 = Radiobutton(container, text="محاضرة وسكشن", variable=course_type_var, value=2)
radio2['font'] = font.Font(size=15)
radio2.pack(anchor = W)

radio2 = Radiobutton(container, text="محاضرة وعملى وتمارين", variable=course_type_var, value=3)
radio2['font'] = font.Font(size=15)
radio2.pack(anchor = W, pady=(0, 20))

button = tk.Button(container, bg='#0052cc', fg='#ffffff', text="اختر الملف",
                   command = lambda: selectFile(course_type_var.get()))
button['font'] = font.Font(size=15)
button.pack()

window.mainloop()
```
